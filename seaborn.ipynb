```python
import seaborn as sns
```

```python
# Basic Plotting with Seaborn
tips = sns.load_dataset('tips')

tips
```

```python
# Create a Scatter Plot

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# Line PLot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total Bill by Size")
plt.xlabel("Size")
plt.ylabel("Total Bill")
```

```python
# Categorical Plots
# Bar Plots

sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar Plot of Total Bill by day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
```

```python
# BoxPlots (For detecting outliers)

sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box Plot Total bill on each day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
```

```python
# Violin Plot( Gives an idea how the distribution of the data is wrt the total bill daywise) 
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
# Histogram

sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
# KDE Plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# Pairplot
# (Gives the reialtionship between each and every feature in the dataset)

sns.pairplot(tips)
```

```python
# Heatmaps

# A Heatmap is a data visualization technique that shows the magnitude of a phenomenon as a color in two dimensions.
# It helps to quickly identify patterns,correlations,or outliers in datasets,especially large ones

# eg :- Lets say you have a data showing student scores across different subjects
#       A heatmap will color each cell based on the scoreâ€”higher scores might be green and lower ones red. This way, you can instantly see whoâ€™s doing well or poorly in which subject.

corr = tips[['total_bill','tip','size']].corr()
corr

sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
# Plotting wrt a particular dataset

import pandas as pd
sales_df = pd.read_csv('100 Sales Records.csv')
sales_df.head()
```

```python
# Plot a total profit by Region

import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y='Total Profit',data=sales_df,estimator=sum)
plt.title('Total Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.show()

# estimator=sum ensures that the bar heights represent the total profit for each region, not the average.
```

```python
# Plot of Total Profit by item type

plt.figure(figsize=(10,6))
sns.barplot(x='Item Type',y='Total Profit',data=sales_df,estimator=sum)
plt.title('Total Revenue by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Revenue')
plt.show()
```
